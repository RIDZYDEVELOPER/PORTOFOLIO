```python
# Program Menghitung Nilai Rata-Rata Siswa

jumlah_siswa = int(input("Masukkan jumlah siswa: "))

data_siswa = []

for i in range(jumlah_siswa):
    print(f"\nSiswa ke-{i+1}")
    nama = input("Nama siswa: ")

    jumlah_nilai = int(input("Berapa banyak nilai yang ingin dimasukkan? "))
    total = 0

    for j in range(jumlah_nilai):
        nilai = float(input(f"Nilai ke-{j+1}: "))
        total += nilai

    rata_rata = total / jumlah_nilai
    data_siswa.append((nama, rata_rata))

print("\n=== HASIL RATA-RATA NILAI SISWA ===")
for nama, rata in data_siswa:
    print(f"Nama: {nama} | Rata-rata nilai: {rata:.2f}")
```

# Bagian Baru

```python
# Menghitung rata-rata 1 siswa dengan 5 nilai

nama = input("Masukkan nama siswa: ")

n1 = float(input("Masukkan nilai 1: "))
n2 = float(input("Masukkan nilai 2: "))
n3 = float(input("Masukkan nilai 3: "))
n4 = float(input("Masukkan nilai 4: "))
n5 = float(input("Masukkan nilai 5: "))

rata_rata = (n1 + n2 + n3 + n4 + n5) / 5

print("\n=== HASIL ===")
print("Nama Siswa :", nama)
print("Rata-rata Nilai :", round(rata_rata, 2))
```
